### 1. 플레이한 주 횟수

한 번 이상 접속한 적 있는 주를 카운트함

```python
play_one = activity_df.groupby('acc_id').size().reset_index(name="play_one")
acc_id_df = play_one[['acc_id']]
play_one.tail()
```

### 2. 주별 플레이한 날짜 수  = 생략

### 3. 총 플레이 시간 합(sum)

```python
sum_cnt = activity_df.groupby('acc_id').sum()
sum_cnt=sum_cnt[['playtime']].tail()
sum_cnt
```

### 4. 경험치 총합_ 경험치의 관점 다시!

솔로 사냥 획득 경험치 + 파티 사냥 획득 경험치  + 퀘스트 획득 경험치 + 경험치 복구 횟수(성당) 

```python
sum_exp = activity_df.loc[:,["acc_id",'solo_exp','party_exp','quest_exp','exp_recovery']].groupby("acc_id").sum().reset_index()
sum_exp.tail()
```

### 5. 나머지 변수 총합

나머지 싸그리

```python
sum_namerge = activity_df.loc[:,['acc_id','npc_kill','boss_monster',
                                 'death','revive','fishing','private_shop',
                                 'game_money_change','enchant_count']].groupby('acc_id').sum().reset_index()
sum_namerge.tail()
```

### 6. 계산된 컬럼 모으기

```python
merge_activity_df = pd.merge(play_one, sum_cnt, how = 'inner', on='acc_id') 
merge_activity_df = pd.merge(merge_activity_df,sum_exp,how = 'inner', on='acc_id'  )
merge_activity_df = pd.merge(merge_activity_df,sum_namerge, how = 'inner', on='acc_id')
merge_activity_df
```
